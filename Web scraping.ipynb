```python
from selenium import webdriver
from webdriver_manager.chrome import  ChromeDriverManager
import time
import warnings
warnings.filterwarnings("ignore")
```

```python
def pause() :
    time.sleep(1) 
```

```python
def search (input_text) :
    driver = webdriver.Chrome(ChromeDriverManager().install())
    driver.get('https://google.com/');
    driver.maximize_window()
    
    search_box = driver.find_element("name", "q")
    search_box.send_keys(f'{input_text} wikipedia')
    search_box.submit()
    
    pause() 
    driver.find_element("xpath", '//*[@id="rso"]/div[1]/div/div/div[1]/div/a/h3/span').click()
    
    text_wikipedia = '' 
    
    for i in range(1,100) :
       try: 
         text = driver.find_element("xpath", f'//*[@id="mw-content-text"]/div[1]/p[{i}]')
        #  print(text.text)
         text_wikipedia +=  text.text + '\n\n\n'
       except : print('Error ' + i ) ; continue
       
       
    driver.quit()
    return text_wikipedia     
```

```python
input_text = input('Enter text to search: ')
print('Your input:', input_text.upper() )
print('-'*30)

text = search(input_text)
print(text)
```
